```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences

```

```python
drive.mount("/content/drive")


# Read the dataset as text
with open("/content/drive/MyDrive/ColabProjects/ner/wikigold/wikigold.conll.txt", 'r', encoding='utf-8') as file:
    dataset_text = file.read()

# Pre-process the dataset to remove trailing "\n" characters
lines = dataset_text.strip().split('\n')
lines = [line.rstrip('\n') for line in lines]

data = []
for line in lines:
    row = line.split(' ')
    data.append(row)

df = pd.DataFrame(data, columns=['Token', 'Entity'])
```

```python
# Remove NA values
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
df["Entity"].value_counts()
```

```python
docstart_positions = df[df['Token'] == '-DOCSTART-'].index
```

```python
sequences = []
start = 0
for position in docstart_positions:
    sequence = df.iloc[start:position]
    sequences.append(sequence)
    start = position + 1
```

```python
sequence_length = 5

tokenizer = Tokenizer()
X = []
y = []
for sequence in sequences:
    tokenizer.fit_on_texts(sequence['Token'])
    X_sequence = tokenizer.texts_to_sequences(sequence['Token'])
    y_sequence = tokenizer.texts_to_sequences(sequence['Entity'])
    X_sequence = pad_sequences(X_sequence, maxlen=sequence_length, padding='post', truncating='post')
    y_sequence = pad_sequences(y_sequence, maxlen=sequence_length, padding='post', truncating='post')
    X.append(X_sequence)
    y.append(y_sequence)
```
